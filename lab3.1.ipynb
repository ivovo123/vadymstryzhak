```python
Так import matplotlib.pyplot as plt
import numpy as np

# Генеруємо дані для функції sin(x)
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)  # Генеруємо 1000 значень x від -2π до 2π
y = np.sin(x)  # Обчислюємо відповідні значення y (y = sin(x))

# Побудова графіка
plt.plot(x, y, label='Функція: y = sin(x)')

# Додавання міток і легенди
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік функції sin(x)')
plt.legend()

# Відображення графіка
plt.grid(True)
plt.show()
```
